#### Min and Max Scaling

In that case we will use a normalization of minimum and maximum: 

$$
x' = \frac{x - min(x)}{max(x) - min(x)}
$$

If we want to normalize the column broccoli, I will take the maximum and minimum value of that column, and then I will look at each individual value to know how for it is for these two values. 

```python
def min_max_scaling(col): 
    min_col = min(col)
    max_col = max(col)
    norm_val = []    
    for val in col: 
        norm_val.append((val - min_col) / (max_col - min_col))
    return norm_val
```

```python
pd.DataFrame({
    'original': df.iloc[:,0], 
    'min_max_scaling': min_max_scaling(df.iloc[:,0])
})
```

```python
# normalise all the columns on a dataset
df_norm = df.apply(min_max_scaling, axis=0).copy()
df_norm
```

```python
# Euclidean (by default)
eucl = pd.DataFrame(pairwise_distances(df_norm), 
                    index=df_norm.index, 
                    columns=df_norm.index)

import seaborn as sns
plt.subplots(figsize=(12, 8))
sns.heatmap(eucl);
```
